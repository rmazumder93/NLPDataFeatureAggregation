```python
def CAGR2(data,start,end):
  stock_data = yf.download(data, start=start, end = end, progress=False)
  df = stock_data.copy()
  df['daily returns'] = df['Adj Close'].pct_change()
  df['cumulative returns'] = (1+ df["daily returns"]).cumprod()
  trading_days = 252 + 251
  n = len(df)/ trading_days
  cagr = ((df["cumulative returns"][-1]**(1/n))-1)
  return cagr

def volatility2(data,start,end):
  stock_data = yf.download(data, start=start, end = end,progress=False)
  df = stock_data.copy()
  df['daily returns'] = df['Adj Close'].pct_change()
  trading_days = 252 + 251
  vol = df['daily returns'].std() * np.sqrt(trading_days)
  return vol

def sharpe(ticker_list,start,end,rf):
  rows = []
  for stocks in ticker_list:
    stock_data = yf.download(stocks, start=start, end = end, progress=False)
    df = stock_data.copy()
    sharpe = (CAGR2(stocks,start,end) - rf) / volatility2(stocks,start,end)
    rows.append([stocks,sharpe])
  df = pd.DataFrame(rows, columns=['All_Tickers','Sharpe ratio'])
  return df
sharpe = sharpe(ticker_list,start,end,.03)
sharpe 
```

```python
#creating volume dataframe and merging them
def vol(stock):
  data = yf.Ticker(stock)
  data = data.history(start= start,end= end, interval='1d')
  data = data[['Volume']].reset_index()
  data['All_Tickers'] = stock
  data.rename(columns={'Date':'post_date'},inplace=True)
  data['post_date']= data['post_date'].astype(str)
  return data
AAPL = vol('AAPL')
TSLA = vol('TSLA')
AMZN = vol('AMZN')
MSFT = vol('MSFT')
GOOGL = vol('GOOGL')
FB = vol('FB')
GOOG = vol('GOOG')
NFLX = vol('NFLX')
F = vol('F')
NVDA = vol('NVDA')
TWTR = vol('TWTR')
A = vol('A')
BABA = vol('BABA')
GS = vol('GS')
AMD = vol('AMD')
T = vol('T')
INTC = vol('INTC')
WMT = vol('WMT')
C = vol('C')
BAC = vol('BAC')
test = pd.concat([AAPL,TSLA,AMZN,MSFT,GOOGL,NFLX,F,NVDA,TWTR,A,BABA,GS,AMD,T,INTC,WMT,C,BAC,GOOG,FB])
test = test.sort_values('post_date').reset_index()
test = test.drop('index', axis=1)
result3 = pd.merge(result2,test,how='left', on =['All_Tickers','post_date'])
result3
```

```python
#creating MA50 dataframe and merging them
def MA50(stock):
  data = yf.Ticker(stock)
  data = data.history(start= '2015-12-31',end= end, interval='1d')
  data = data[['Volume']].reset_index()
  data['All_Tickers'] = stock
  data.rename(columns={'Date':'post_date'},inplace=True)
  data['post_date']= data['post_date'].astype(str)
  data['50MA'] = data['Volume'].rolling(50).mean()
  data = data[(data['post_date'] >= "2017-01-01") & (data['post_date'] <= "2018-12-31")]
  return data
AAPL = MA50('AAPL')
TSLA = MA50('TSLA')
AMZN = MA50('AMZN')
MSFT = MA50('MSFT')
GOOGL = MA50('GOOGL')
FB = MA50('FB')
GOOG = MA50('GOOG')
NFLX = MA50('NFLX')
F = MA50('F')
NVDA = MA50('NVDA')
TWTR = MA50('TWTR')
A = MA50('A')
BABA = MA50('BABA')
GS = MA50('GS')
AMD = MA50('AMD')
T = MA50('T')
INTC = MA50('INTC')
WMT = MA50('WMT')
C = MA50('C')
BAC = MA50('BAC')
testMA = pd.concat([AAPL,TSLA,AMZN,MSFT,GOOGL,NFLX,F,NVDA,TWTR,A,BABA,GS,AMD,T,INTC,WMT,C,BAC,GOOG,FB])
testMA = testMA.sort_values('post_date').reset_index()
testMA = testMA.drop('index', axis=1)
testMA = testMA.drop('Volume', axis=1)
testMA['50MA'] =round(testMA['50MA'])
result4 = pd.merge(result3,testMA,how='left', on =['All_Tickers','post_date'])
result4
```

```python
#creating MA200 dataframe and merging them
def MA200(stock):
  data = yf.Ticker(stock)
  data = data.history(start= '2015-12-31',end= end, interval='1d')
  data = data[['Volume']].reset_index()
  data['All_Tickers'] = stock
  data.rename(columns={'Date':'post_date'},inplace=True)
  data['post_date']= data['post_date'].astype(str)
  data['200MA'] = data['Volume'].rolling(200).mean()
  data = data[(data['post_date'] >= "2017-01-01") & (data['post_date'] <= "2018-12-31")]
  return data
AAPL = MA200('AAPL')
TSLA = MA200('TSLA')
AMZN = MA200('AMZN')
MSFT = MA200('MSFT')
GOOGL = MA200('GOOGL')
FB = MA200('FB')
GOOG = MA200('GOOG')
NFLX = MA200('NFLX')
F = MA200('F')
NVDA = MA200('NVDA')
TWTR = MA200('TWTR')
A = MA200('A')
BABA = MA200('BABA')
GS = MA200('GS')
AMD = MA200('AMD')
T = MA200('T')
INTC = MA200('INTC')
WMT = MA200('WMT')
C = MA200('C')
BAC = MA200('BAC')
testMA2 = pd.concat([AAPL,TSLA,AMZN,MSFT,GOOGL,NFLX,F,NVDA,TWTR,A,BABA,GS,AMD,T,INTC,WMT,C,BAC,GOOG,FB])
testMA2 = testMA2.sort_values('post_date').reset_index()
testMA2 = testMA2.drop('index', axis=1)
testMA2 = testMA2.drop('Volume', axis=1)
testMA2['200MA'] =round(testMA2['200MA'])
result5 = pd.merge(result4,testMA2,how='left', on =['All_Tickers','post_date'])
result5
```

```python
#creating volatility dataframe and merging them
def volatility(ticker_list,start,end):
  rows = []
  for stocks in ticker_list:
    stock_data = yf.download(stocks, start=start, end = end,progress=False)
    df = stock_data.copy()
    df['daily returns'] = df['Adj Close'].pct_change()
    trading_days = 252 + 251
    vol = round((df['daily returns'].std() * np.sqrt(trading_days))*100)
    vol = str(vol)+'%'
    rows.append([stocks,vol])
  df = pd.DataFrame(rows, columns=['All_Tickers','Two Years Volatility'])
  return df

volatility = volatility(ticker_list,start,end)
result6 = pd.merge(result5,volatility,how='left', on =['All_Tickers'])
result6
```

```python
# Creating Volume change % for 200MA and 50MA
result7 = result6.copy()
result7 = result7.fillna(0)
result7['VolumeChange_50MA'] = round(((result7['Volume'] - result7['50MA'])/result7['50MA'])*100)
result7 = result7.fillna(0)
result7
result7['VolumeChange_200MA'] = round(((result7['Volume'] - result7['200MA'])/result7['200MA'])*100)
result7 = result7.fillna(0)
result7['VolumeChange_200MA'] = pd.Series(["{0:.2f}%".format(val) for val in result7['VolumeChange_200MA']])
```
